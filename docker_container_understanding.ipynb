##### Docker Container Understanding

We install docker on our machine first like .exe  docker file for windows.
After this we install container on docker using command docker run 

"docker run" is a command to create container. In the below command what we are doing is pulling docker image busy box from docker hub which is a registry of doker images.



docker info 
docker run busybox:1.24 echo "hello world"
docker images
docker run busybox:1.24 ls -l 
docker ps -a


I created a docker file which is a text file of instructions. 
From docker file we create docker image.
Docker image is used to create containers.

### Difference between VM and containers

We have physical server on top of it we have OS running and then we install hypervisor that gives us the capability to install multiple VMs . VM has its own OS, disk space , ram etc.Virtualization happens at hardware level.

However containers are light weight which has all the libraries and binaries needed to install the application.so we deply containers on any machine which has docker installed and then we run our application on different platform using containers. So in containers virtualization happens at OS level.
